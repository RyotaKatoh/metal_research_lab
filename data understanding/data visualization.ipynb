```python
import json
```

```python
with open('./bronze/bronze_se_2d.json', 'r') as f:
    bronze_se_2d = json.load(f)
```

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
```

```python
output_notebook()
```

```python
import os
```

```python
p = figure(width=800, height=800, title='bronze se 2D')
```

```python
for item in bronze_se_2d['elements']:
    if item['img_path'] == '':
        p.circle([item['values'][0]], [item['values'][1]], color='#43ab32', fill_alpha=0.4, size=5)
    else:
        p.image_url(url=[os.path.join('./bronze/', item['img_path'])], x=item['values'][0], y=item['values'][1], w=0.05, h=0.1)
        p.circle([item['values'][0]], [item['values'][1]], color='#ca25ff', fill_alpha=0.4, size=5)

show(p)
```
